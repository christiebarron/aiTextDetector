## Read the data into the excel document

```python
import pandas as pd
#save all the ai essays to a pandas dataframe ######
working_directory = "../rawData/aiEssays/"#"primary/rawData/aiEssays/"
clean_data_directory = "../cleanData/" #


text_list = [] #create a list to add text files to.
ai_llm = [] #a list of the large language model used
eid = [] #essay prompt id
row_id = [] #final number
files = os.listdir(working_directory) #get a list of all file names within the directory
test1 = "eid8_text-davinci-003_1352.txt"
test2 = "8_gpt2_5059.txt"
test3 = 'eid21-the-good-ai_2909.txt'
```

```python
#extract relevant information from ai-generated text files
for f in files:
    filename = f'{working_directory}{f}' #save the filename

    with open(filename, 'r') as f: #open filename and save the text in it
        text_list.append(f.read())
    
    #append metadata saved in the filename (llm used, essay prompt, and row id)
    if "the-good-ai" in filename:
        filename = f"{filename.split('-')[0]}_the-good-ai_{filename.split('_')[1]}"
        
    try:
        ai_llm.append(filename.split("_")[1]) #second element is the LLM used to generate the text

        if (filename.split("_")[0].startswith('e') == False):
            eid.append(filename.split('_')[0]) #for essays that weren't prefaced with "eid", just append the number
        else:
            eid.append(filename.split("_")[0].split("d")[1]) #if essays were prefaced w/ "eid", separate then just append the number

        row_id.append(filename.split("_")[2].split(".")[0]) #append the row id, which is 3rd element.
    except:
        print(f"error with {filename}")
    
#save all extracted text to a pandas dataframe, then excel file.
df = pd.DataFrame({"row_id" : row_id, "essay_id" : eid, "ai_llm": ai_llm, 'ai_essay': text_list})
df.to_excel("{clean_data_directory}aiGenerated.xlsx")
```
