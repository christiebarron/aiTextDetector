```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('3aMergedASAP.xlsx')

# Filter the rows based on the values in the "ai_lim" column
human_df = df[df['ai_llm'] == 'human-generated']
ai_df = df[df['ai_llm'] == 'text-davinci-003']

# Save the filtered rows in separate Excel files
human_df.to_excel('human-generated.xlsx', index=False)
ai_df.to_excel('ai-generated.xlsx', index=False)
```

```python
ai_df
```

```python
import pandas as pd
import string

# Remove punctuation from the text data
translator = str.maketrans('', '', string.punctuation)

human_df['essay'] = human_df['essay'].apply(lambda x: x.translate(translator))
ai_df['essay'] = ai_df['essay'].apply(lambda x: x.translate(translator))

# Save the cleaned data in separate Excel files
human_df.to_excel('human-generated-cleaned.xlsx', index=False)
ai_df.to_excel('ai-generated-cleaned.xlsx', index=False)

```

```python
import re
import string

# Define the list of uppercase words to remove
uppercase_words = ["CAPS1", "CAPS2","CAPS3","CAPS8", "ORGANIZATION1", "ORGANIZATION2", "LOCATION1", "LOCATION2"]

# Define a regular expression pattern that matches any of the uppercase words
pattern = re.compile("|".join(uppercase_words))

# Remove the uppercase words from the "essay" column
human_df["essay"] = human_df["essay"].apply(lambda x: pattern.sub("", x))

# Remove punctuation from the "essay" column
ai_df["essay"] = ai_df["essay"].apply(lambda x: x.translate(str.maketrans("", "", string.punctuation)))

# Save the cleaned essays to a new Excel file
df.to_excel("clean_essays.xlsx", index=False) 
```

```python
import numpy as np
from scipy import stats

essay_len = human_df['essay'].apply(len)
mean_len = essay_len.mean()
median_len = essay_len.median()
mode_len = stats.mode(essay_len)[0][0]
quartile_25 = np.percentile(essay_len, 25)
quartile_50 = np.percentile(essay_len, 50)
quartile_75 = np.percentile(essay_len, 75)
std_dev = ai_df["essay"].str.len().std()

# Print the descriptive statistics
print(f"Mean length: {mean_len:.2f}")
print(f"Median length: {median_len}")
print(f"Mode length: {mode_len}")
print(f"First quartile (25%): {quartile_25}")
print(f"Second quartile (50%): {quartile_50}")
print(f"Third quartile (75%): {quartile_75}")
print(f"Standard deviation: {std_dev:.2f}")
```

```python
import matplotlib.pyplot as plt

# Create a box plot of the essay lengths
plt.boxplot(human_df["essay"].str.len())

# Add labels and title to the plot
plt.xlabel("Essays")
plt.ylabel("Length")
plt.title("Box plot of essay lengths")

# Show the plot
plt.show()
```
