```python
# # Scrape Multiple Pages
from splinter import Browser
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as soup
import time

browser=Browser('chrome')
```

```python
# Visit the URL

base_url = 'https://studyfy.com/essay-samples'
browser.visit(base_url)

# Parse the HTML
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract and store the links for each of the essay 
urls = []
sub_urls =[]

for cat in soup.find_all("ul", class_='category__list w-list-unstyled'):
    
    for link in soup.find_all("a", class_='category__link'):
        links=link.get("href")
        urls.append(links)
        
    for urls in soup.find_all("a", class_='post-preview d-block'):
            sublinks=urls.get("href")
            sub_urls.append(sublinks)

```

```python
import numpy as np

my_array = np.asarray(urls)
```

```python
sub_urls=[]

for i in range(len(my_array)):
    print(f"working on page {i}")
    combined_url = f"{my_array[i]}"
    browser.visit(combined_url)
    time.sleep(4)
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes=quote_soup.find_all("a", class_='post-preview d-block')
    for quote in quotes:
        sub_urls.append(quote.get("href"))
```
